```python
from sentence_transformers import SentenceTransformer, util
from scipy import spatial
sentences = ["This is a Norwegian boy", "Dette er en norsk gutt"]

model = SentenceTransformer('NbAiLab/nb-sbert-base')
```

```python

def check_cosine(sentences):
    embeddings = model.encode(sentences)
    print(embeddings)
    scores = []
    scores.append(1)

    for embedding in embeddings[1:]:
        #cosine_scores = util.cos_sim(embeddings[0],embedding)
        scipy_cosine_scores = 1 - spatial.distance.cosine(embeddings[0],embedding)
        scores.append(scipy_cosine_scores)
    return scores
```

```python
with open('82060599.txt') as f:
    text = f.readlines()
```

```python
sentences = text[0].split('.')
```

```python
scores = check_cosine(sentences)
```
